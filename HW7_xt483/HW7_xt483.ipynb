```python
fig1 = plt.figure()
# Make a legend for specific lines.
import matplotlib.pyplot as plt
import numpy as np


t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)
plt.xkcd()
# note that plot returns a list of lines.  The "l1, = plot" usage
# extracts the first element of the list into l1 using tuple
# unpacking.  So l1 is a Line2D instance, not a sequence of lines
l1, = plt.plot(t2, np.exp(-t2),'b-', alpha = 0.5)
l2,l3 = plt.plot(t2, np.sin(2 * np.pi * t2), '--bo', t1, np.log(1 + t1), 'r-', alpha =0.5)
l4, = plt.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), '-ro', alpha = 0.5)
plt.legend((l1, l2,l3,l4), ('exp','oscillatory', 'log','damped'), loc='best', shadow=True)
plt.xlabel('time/s')
plt.ylabel('volts/v')
plt.title('Damped oscillation')
plt.ylim([-1.5,1.5])
plt.show()
```
